### Feature Selection

##### Feature selection is the process of extracting the most relevant features from a dataset(removing unwanted columns)

```python
#checking all columns 
app.columns
```

```python
#checking total columns and rows
app.shape
```

```python
#getting total null values based on each columns
app.isnull().sum().sort_values()
```

```python
#making list into dataframe and renaming the columns
msng_info = pd.DataFrame(app.isnull().sum().sort_values()).reset_index()
msng_info.rename(columns = {'index' : 'col_name', 0 : 'null_counts'}, inplace = True)
msng_info.head()
```

```python
#getting percentage of null values and exporting dataset into excel for reference
msng_info['msng_pct'] = msng_info['null_counts']/app.shape[0]*100
#msng_info.to_excel(r'D:\Python Projects\Loan_defaulter_data\msng_info.xlsx', index = False)
msng_info
```

```python
#droping more then 40% null values
#The rule says the columns which contain more then 40% of null values are not good for analysis and imputation
msng_col = msng_info[msng_info['msng_pct']>40]['col_name'].to_list()
app_msng_rmvd = app.drop(labels = msng_col, axis = 1)
app_msng_rmvd.shape
```

```python
app_msng_rmvd.head()
```

```python
#checkinf flag columns what they are atually in those what are usefull for analysis
flag_col = []
for col in app_msng_rmvd.columns:
    if col.startswith("FLAG_"):
        flag_col.append(col)

len(flag_col)
```

```python
#comparing flag columns with target
flag_trgt_col = app_msng_rmvd[flag_col+['TARGET']]
flag_trgt_col.head()
```

```python
#most of the customers who hasn't submitted FLAG_DOCUMENT_19, most of them are terned as non defaulters.
sns.countplot(data = flag_trgt_col, x = 'FLAG_DOCUMENT_19' , hue = 'TARGET')
plt.show()
```

```python
'''most of the customers not submitted documents and they turned into non defaulters too, 
that is why we can consider the flag columns are not best for analysis'''
plt.figure(figsize = (25,25))
for i, col in enumerate(flag_col):
    plt.subplot(7,4, i+1)
    sns.countplot(data = flag_trgt_col, x = col, hue = 'TARGET')
```

```python
#making data frame for perticular flag columns
flag_corr = ['FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'FLAG_MOBIL', 'FLAG_EMP_PHONE', 'FLAG_WORK_PHONE', 'FLAG_CONT_MOBILE', 'FLAG_PHONE', 'FLAG_EMAIL', 'TARGET']
flag_corr_df = app_msng_rmvd[flag_corr]
flag_corr_df.head()
```

```python
#.loc[;, used because of warning and replaced 'N' and 'Y' to 0 and 1
flag_corr_df.loc[:,'FLAG_OWN_CAR'] = flag_corr_df['FLAG_OWN_CAR'].replace({'N' : 0, 'Y' : 1})
flag_corr_df.loc[:,'FLAG_OWN_REALTY'] = flag_corr_df['FLAG_OWN_REALTY'].replace({'N' : 0, 'Y' : 1})
```

```python
#correlation betwwen target and all the flag columns
plt.figure(figsize = (8, 6))
corr_df = round(flag_corr_df.corr(),2)
sns.heatmap(corr_df, cmap = 'coolwarm', linewidth = 5, annot = True)
plt.show()
```
